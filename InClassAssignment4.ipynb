# 2. Sorting text into a dictionary 

- Define a function which is passed a string of text as a parameter. (use an input statement)

- The function returns a dictionary with keys which are integers and corresponding values which are lists of unique words. 

- The list of words corresponding to a key contains all the unique words from the text that have a length equal to the key value. 

- The corresponding lists of unique words should be in sorted alphabetical order.



For example, for the following text strings:

text1 = "May your coffee be strong and your Monday be short"



text2 = 'why does someone believe you when you say there are four billion stars but they have to check when you say the paint is wet'

text 1 prints:

- 2 : ['be']
- 3 : ['May', 'and']
- 4 : ['your']
- 5 : ['short']
- 6 : ['Monday', 'coffee', 'strong']

text2 prints:

- 2 : ['is', 'to']
- 3 : ['are', 'but', 'say', 'the', 'wet', 'why', 'you']
- 4 : ['does', 'four', 'have', 'they', 'when']
- 5 : ['check', 'paint', 'stars', 'there']
- 7 : ['believe', 'billion', 'someone']

```python
def sort_text(text: str):
    words = text.split(" ")
    sorted_words = sorted(words, key=len)

    result = {}
    for word in sorted_words:
        if len(word) in result.keys():
            result[len(word)].append(word)
        else:
            result[len(word)] = [word]

    return result



text1 = "May your coffee be strong and your Monday be short"
print (sort_text(text1))
```


# 3. State Trivia Facts


## Input:
- Prompt the user for a State, or q to quit
- Prompt the user to enter facts about the state, or q to quit
- User can enter 0 to n States & Facts

## Output
### Use a dictionary to print back the state and facts, for example:


- Some facts about the state of Ohio:
    - Nicknamed the Buckeye State
    - State Capital is Columbus


- Some facts about the state of Virginia:
    - Home of Virginia Commonwealth University


```python

while True:
    state = input("Enter a State or q to quit:")
    if state == 'q':
        break

    facts = []
    while True:
        fact = input(f'Enter a fact about {state} or q to quit')
        facts.append(fact)
        if fact =='q':
            break

    print(state)
    for fact in facts:
        print(fact)
```

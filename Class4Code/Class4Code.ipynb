## AND (&)  VS OR (|) Conditional Operators

```python
pets = ["Winnie", "Lady", "Pancake"]
```

```python
if ("Winnie" and "Lady" in pets):
    print("This is TRUE, Winnie & Lady are BOTH in your pet list")
else:
    print("Winnie AND Lady are not in your pet list")
```

```python
if ("Winnie" or "Samson" in pets):
    print("Winnie is in the list, but Samson is NOT. We only needed ONE present to make the conditional true")
else:
    print("Neither Winnie nor Samson is in the pets list.")
```

# Loops

```python
candies = ["Skittles", "Air Heads", "Starburts"]
for i in candies:
    print(i)
```

## LEARN HOW TO CLEAR KERNEL/STOP CODE FROM RUNNING.

```python
i = 0
while (i <= len(candies)):
    print(i)
```

```python
i = 0
while (i < len(candies)):
    print(i)
    i = i + 1
```

```python
i = 0
while (i <= len(candies)):
    print(i)
    i = i + 1
```

## While Loop Breaks

```python
prompt = "\nTell me something, and I will repeat it back to you:"
prompt += "\nEnter 'quit' to end the program. "
message = ""
while message != 'quit':
    message = input(prompt)
    if message != 'quit':
        print(message)
```
